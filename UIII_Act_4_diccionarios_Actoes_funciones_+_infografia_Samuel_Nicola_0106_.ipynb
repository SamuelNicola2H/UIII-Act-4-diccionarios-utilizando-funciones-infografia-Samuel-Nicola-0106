### Diccionario Actores

```python
# Inicialización de la lista de actores
agencia_actores = []

def agregar_actor():
    actor = {}
    print("\n--- Agregar nuevo actor ---")
    actor["nombre"] = input("Nombre: ")
    actor["obras"] = input("Obras realizadas: ")
    actor["premios"] = input("Premios: ")
    actor["habilidades"] = input("Habilidades especiales: ")
    actor["edad"] = input("Rango de edad: ")
    actor["ultimo"] = input("Último proyecto: ")
    agencia_actores.append(actor)
    listar_actores()

def listar_actores():
    print("\n" + "="*110)
    print(f"{'NOMBRE':<15} | {'OBRAS':<15} | {'PREMIOS':<15} | {'HABILIDADES':<15} | {'EDAD':<10} | {'ÚLTIMO PROYECTO'}")
    print("-" * 110)
    for actor in agencia_actores:
        print(f"{actor['nombre']:<15} | {actor['obras']:<15} | {actor['premios']:<15} | {actor['habilidades']:<15} | {actor['edad']:<10} | {actor['ultimo']}")
    print("="*110)

def buscar_actor():
    nombre = input("Nombre a buscar: ")
    for actor in agencia_actores:
        if actor["nombre"] == nombre:
            for key, value in actor.items():
                print(f"{key.upper()}: {value}")
    listar_actores()

def borrar_actor():
    nombre = input("Nombre del actor a borrar: ")
    for i in range(len(agencia_actores)):
        if agencia_actores[i]["nombre"] == nombre:
            agencia_actores.pop(i)
            break
    listar_actores()

def actualizar_actor():
    nombre = input("Nombre del actor a actualizar: ")
    for actor in agencia_actores:
        if actor["nombre"] == nombre:
            actor["obras"] = input("Nuevas obras: ")
            actor["premios"] = input("Nuevos premios: ")
    listar_actores()

# Menú principal
while True:
    print("\n--- MENÚ AGENCIA ---")
    print("1. Agregar actor\n2. Listar actores\n3. Buscar actor\n4. Borrar actor\n5. Actualizar actor\n6. Salir")
    opcion = input("Elige una opción: ")

    if opcion == "1": agregar_actor()
    elif opcion == "2": listar_actores()
    elif opcion == "3": buscar_actor()
    elif opcion == "4": borrar_actor()
    elif opcion == "5": actualizar_actor()
    elif opcion == "6": break
```
